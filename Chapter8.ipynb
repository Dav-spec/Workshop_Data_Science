```python
img = cv2.imread('boxes.jpg')
gray = np.float32(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
dst = cv2.cornerHarris(gray,10,1,0.1)
plt.imshow(dst, cmap='gray')
plt.savefig('Vision 3.png', dpi=350)
```

```python
img = cv2.imread('boxes.jpg')
gray = np.float32(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
dst = cv2.cornerHarris(gray,10,1,0.1)
for i in [0.1, 0.01, 0.001, 0.0001]:
    img[dst>i*dst.max()]=[255,0,0]
    plt.title('Harris Corner Detection, Th = {}'.format(i))
    plt.imshow(img)
    plt.savefig('Vision 4_{}.png'.format(i), dpi=350)
    plt.show()
```

```python
img = cv2.imread('boxes.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

n_features = 120

corners = cv2.goodFeaturesToTrack(gray,n_features,0.01,10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),10,255,-1)

plt.imshow(img)
plt.title('{} features'.format(n_features))
plt.savefig('Vision 5_exc_{}.png'.format(n_features), dpi=350)
plt.show()
```
